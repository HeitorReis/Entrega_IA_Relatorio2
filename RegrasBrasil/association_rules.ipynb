```python
# Checking the existence of non-binary values
nonbinary_columns = []

for index, value in enumerate(data.values[0]):
  if value not in ['Sim', 'Não', '-', 'Não sabe']:
    
    this_column = data.columns[index]
    if this_column not in nonbinary_columns:
      nonbinary_columns.append([data.columns[index], index])

for item in nonbinary_columns:
  print( str(item) )
del nonbinary_columns
```

```python
# I need to adapt this dataset so that I can analyse it using the AI.
# The data is using binary inputs in order to track the occurrences of
# risk management measures, so I have to replace all 'Yes's with the name
# of the column and delete the options with 'No' or '-'.

relation_dataset = []

for mun_index, municipality in enumerate(data.values): # For every municipality
  
  relation_dataset.append(list()) # Allocate the municipality space in memory 
  
  for att_index, attribute in enumerate(municipality): # For every attribute
    
    this_att_name = data.columns[att_index] # Take the att name
    
    if attribute == 'Sim': # If the attribute is existent
      relation_dataset[mun_index].append(this_att_name) # Add to the new dataset
    
    elif (attribute != 'Não') and (attribute != '-') and (attribute != 'Não sabe') and (attribute != 'Recusa') and (type(attribute) is not int): # If isn't binary
      relation_dataset[mun_index].append(attribute) # Register att the new dataset

relation_dataset[0]
```

```python
# Using mlxtend to get the frequent itemsets (combination of items)

te = TransactionEncoder() # Initialize the class
te_ary = te.fit(main_dataset.users).transform(main_dataset.users) # Fit data
df = pd.DataFrame(te_ary, columns=te.columns_) # View using pandas

min_support = 0.05

frequent_itemsets = fpgrowth(df, min_support=min_support, use_colnames=True) # Training AI

frequent_itemsets.shape
```

```python
filename = f'generated/brasil_frequent_itemsets'
with open(filename+'.csv', 'w') as output_handle: # Save as .csv for visualization
  frequent_itemsets.to_csv(filename+'.csv', columns=frequent_itemsets.columns)
with open(filename+'.xlsx', 'w') as output_handle: # Save as .csv for visualization
  frequent_itemsets.to_excel(filename+'.xlsx', columns=frequent_itemsets.columns)
```

```python
frequent_itemsets.head() # View
```

```python
# Generating associations only with lift higher than 3
raw_rules = association_rules(frequent_itemsets, metric="lift", min_threshold=3)
raw_rules.shape
```

```python
# Filtering only zhangs > 0.6 and lift > 1.1
# (Filtering important relations and correlated data only)
rules = raw_rules[
    (raw_rules['zhangs_metric'] > 0.95)
    & (raw_rules['lift'] > 3)
    #& (raw_rules['confidence'] > 0.6)
    ]

print(rules.shape)
rules.head() # View
```
