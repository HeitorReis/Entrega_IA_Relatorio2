```python
# Using mlxtend to get the frequent item sets (combination of items)

region_rules = []
region_frequent_itemsets = []

for i, this_region_dataset in enumerate(region_dataset):
  te = TransactionEncoder() # Initialize the class
  te_ary = te.fit(this_region_dataset).transform(this_region_dataset) # Fit data
  df = pd.DataFrame(te_ary, columns=te.columns_) # View using pandas

  min_support = 0.4 # Define min_support

  aux1 = apriori(df, min_support=min_support, use_colnames=True)
  region_frequent_itemsets.append(aux1) # Training AI
  
  aux2 = association_rules(region_frequent_itemsets[i], metric="zhangs_metric", min_threshold=0.6)
  region_rules.append(aux2)
  
  filename = f'region-rules/region{i+1}-rules.csv'
  with open(filename, 'w') as output_handle: # Save as .csv for visualization
    region_rules[i].to_csv(filename, columns=region_rules[i].columns)
    
  del te, te_ary, df, aux1, aux2
```
