```python
relation_dataset_1 = [] # Norte
relation_dataset_2 = [] # Nordeste
relation_dataset_3 = [] # Sudeste
relation_dataset_4 = [] # Sul
relation_dataset_5 = [] # Centro-Oeste


for sample_index, sample in enumerate(data.values):
    if '1' in data.loc[sample_index, 'Regiao']:
        relation_dataset_1.append(data.values[sample_index])
    elif '2' in data.loc[sample_index, 'Regiao']:
        relation_dataset_2.append(data.values[sample_index])
    elif '3' in data.loc[sample_index, 'Regiao']:
        relation_dataset_3.append(data.values[sample_index])
    elif '4' in data.loc[sample_index, 'Regiao']:
        relation_dataset_4.append(data.values[sample_index])
    else:
        relation_dataset_5.append(data.values[sample_index])

relation_dataset_4[0]
```

```python
all_datasets = [
    relation_dataset_1,
    relation_dataset_2, 
    relation_dataset_3,
    relation_dataset_4, 
    relation_dataset_5
    ]
```

```python
all_dframes = []
for dataset in all_datasets:
    all_dframes.append(pd.DataFrame(dataset, columns=data.columns))
```

```python
all_dframes[0].head()
```

```python
region_occurrences_list = []

for dframe_index, this_dframe in enumerate(all_dframes):
    
    new_dataset = []

    for mun_index, municipality in enumerate(this_dframe.values): # For every municipality
    
        new_dataset.append(list()) # Allocate the municipality space in memory 
        
        for att_index, attribute in enumerate(municipality): # For every attribute
            
            this_att_name = this_dframe.columns[att_index] # Take the att name
            
            if attribute == 'Sim': # If the attribute is existent
                new_dataset[mun_index].append(this_att_name) # Add to the new dataset
            
            elif (attribute != 'Não') and (attribute != '-') and (attribute != 'Não sabe') and (attribute != 'Recusa') and (type(attribute) is not int): # If isn't binary
                new_dataset[mun_index].append(attribute) # Register to the new dataset
    
    region_occurrences_list.append(new_dataset)

region_occurrences_list[0][0]
```

```python
# AI training
frequent_itemsets_list = []

for dataset_index, this_dataset in enumerate(dataset_list):
    # Using mlxtend to get the frequent itemsets (combination of items)

    te = TransactionEncoder() # Initialize the class
    te_ary = te.fit(this_dataset.users).transform(this_dataset.users) # Fit data
    df = pd.DataFrame(te_ary, columns=te.columns_) # View using pandas

    min_support = 0.05 # Define min_support (set to 0.2 because the 'Regiao' att is approximately 20%, but this isn't precise)

    frequent_itemsets_list.append( fpgrowth(df, min_support=min_support, use_colnames=True) )# Training AI
    print(frequent_itemsets_list[dataset_index].shape)
```

```python
# Saving
for i, rule in enumerate(frequent_itemsets_list):
    filename = f'generated/support_tables/support_list_{i+1}.xlsx'
    with open(filename, 'w') as output_handle: # Save as .csv for visualization
        rule.to_excel(filename, columns=rule.columns)
```

```python
raw_rules_list_V1 = []
min_threshold = 1.0

for region_index, frequent_itemsets in enumerate(frequent_itemsets_list):
    raw_rules_list_V1.append( association_rules(frequent_itemsets, metric='zhangs_metric', min_threshold=min_threshold) )
    raw_rules_list_V1[region_index] = raw_rules_list_V1[region_index][raw_rules_list_V1[region_index]['confidence'] < 0.30]
    print(f'{region_index+1}:',raw_rules_list_V1[region_index].shape)
```

```python
raw_rules_list_V2 = []
min_threshold = 10

for region_index, frequent_itemsets in enumerate(frequent_itemsets_list):
    if raw_rules_list_V1[region_index].shape[0] > 100:
        min_threshold += -6
        raw_rules_list_V2.append(raw_rules_list_V1[region_index][raw_rules_list_V1[region_index]['lift'] > min_threshold])
        print(f'{region_index+1}:',raw_rules_list_V2[region_index].shape)
    else:
        raw_rules_list_V2.append(raw_rules_list_V1[region_index])
```
